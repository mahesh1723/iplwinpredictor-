```python
#total_runs_x is total score
#total_runs_y is score per ball
```

```python
#renaming the columns
delivery_df.rename(columns={'total_runs_x':'total_runs', 'total_runs_y':'Ball_score'}, inplace=True)
```

```python
#new column which will say what is score per ball 
#added score upto that point of time
```

```python
delivery_df['Score'] = delivery_df[['match_id','Ball_score']].groupby('match_id').cumsum()['Ball_score']
```

```python
delivery_df
```

```python
#new column which has target which is getting revised
delivery_df['target_left'] = (delivery_df['total_runs']+1) - delivery_df['Score']
```

```python
delivery_df
```

```python
# 1 over we have 6 balls 
#ipl has 20 over match 20 * 6 =120 balls
delivery_df['Remaining_Balls'] = (120 - ((delivery_df['over']-1)*6 + delivery_df['ball']))
```

```python
delivery_df
```

```python
#HOW MANY wickets are left out
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna('0')
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == '0' else '1').astype('int64')
```

```python
delivery_df[['match_id','player_dismissed']]
```

```python
delivery_df
```

```python
delivery_df['wickets'] = delivery_df[['match_id','player_dismissed']].groupby('match_id').cumsum()['player_dismissed'].values
```

```python
delivery_df
```

```python
#how many wickets are left out
delivery_df['wickets'] = 10 - delivery_df['wickets']
```

```python
delivery_df
```

Current run rate

$$current \ run \ rate = \frac{Runs \ scored} {Number \ of \ Overs \ Completed} $$

```python
delivery_df['crr'] =(delivery_df['Score'])*6/(120 - delivery_df['Remaining_Balls'])
```

```python
delivery_df
```

Required Run Rate

$$required \ run \ rate = \frac{Remaining \ score } {Number \ of \ Overs \ left} $$

```python
delivery_df['rrr'] =(delivery_df['target_left'])*6/(delivery_df['Remaining_Balls'])
```

```python
delivery_df
```

```python
#xreating a columns as result where 1 means won and 0 means loss
def result(row):
    if row['batting_team'] == row['winner']:
        return 1
    else:
        return 0
```

```python
delivery_df['result'] = delivery_df.apply(result, axis=1)
```

```python
model_df = delivery_df[['batting_team', 'bowling_team', 'city', 'Score', 'wickets','Remaining_Balls','target_left','crr','rrr','result']]
```

```python
model_df = model_df[model_df['Remaining_Balls']!=0]
```

```python
#splitting the data into feature and target
X = model_df.iloc[:,:-1]
y = model_df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
#splitting the data to train and test 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=2)
```

```python
X_test
```

```python
y_test
```

```python
#label encoding 
#one hot encoding
#transform non numerical data to numerical data
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'),['batting_team','bowling_team','city'])],remainder = 'passthrough')
```

```python
trf
```

```python
#building the mdoel
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2', LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_prediction = pipe.predict(X_test)
```

```python
y_prediction
```

```python
#accuracy score
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_prediction)
```
